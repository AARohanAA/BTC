```
from bs4 import BeautifulSoup
import requests 
import time
import datetime
import matplotlib.pyplot as plt


def get_crypto_price(coin):

    url = "https://www.google.com/search?q="+coin+"+price usd"
    HTML = requests.get(url) 
    soup = BeautifulSoup(HTML.text, 'html.parser') 
    text = soup.find("div", attrs={'class':'BNeawe iBp4i AP7Wnd'}).find("div", attrs={'class':'BNeawe iBp4i AP7Wnd'}).text
    return text
    
def conv(price):
    price = price.split()
    price = price[0]
    x = 0
    for i in range(len(price)):
        if price[i] == ',':
           x = i
    return price[:x] + price[x+1:]
def mode1():
  crypto = input('Enter Crypto(bitcoin, ethereum): ')
  times = int(input('Enter time for observation: '))
  last_price = -1
  x = []
  y = []
  v = 1
  end_time = time.time() +  times
  min_price = 0
  max_price = 0
  while time.time() < end_time:
      
    price = get_crypto_price(crypto)
    print(crypto+' price(dollars): ', conv(price))
    last_price = price 
    y.append(float(conv(price)))

    x.append(v)
    v+=1
    
    if float(conv(price)) <= min(y):
        min_price = price
        min_time = datetime.datetime.now().time()
    if float(conv(price)) >= max(y):
        max_price = price
        max_time = datetime.datetime.now().time()
        
  print('Change: ', float(y[-1])-float(y[0]))
  print('Percent change: ', (float(y[-1])-float(y[0]))/float(y[0]))
  print('Min price, time: ', min_price, min_time)
  print('Max price, time: ' , max_price, max_time)
  plt.plot(x,y)
  plt.show()
  again = input('Back to main menu?(y/n)')
  if again == 'y':
     main()
    
def mode2():
    times = int(input('Enter time for observation: '))
    last_price_btc = -1
    last_price_eth = -1
    ybtc = []
    yeth =[]
    x = []
    v = 1
    min_pricebtc = 0
    max_pricebtc = 0
    min_priceeth = 0
    max_priceeth = 0
    cryptobtc = 'bitcoin'
    cryptoeth = 'ethereum'
    end_time = time.time() +  times
    while time.time() < end_time: 
        pricebtc = get_crypto_price(cryptobtc)
        priceeth = get_crypto_price(cryptoeth)
        print(cryptobtc+' price(dollars): ', conv(pricebtc))
        print(cryptoeth+' price(dollars): ', conv(priceeth))
        last_pricebtc = pricebtc
        last_priceeth = priceeth
        ybtc.append(float(conv(pricebtc)))
        yeth.append(float(conv(priceeth)))

        x.append(v)
        v+=1
        if float(conv(pricebtc)) <= min(ybtc):
            min_pricebtc = pricebtc
            min_timebtc = datetime.datetime.now().time()
        if float(conv(pricebtc)) >= max(ybtc):
            max_pricebtc = pricebtc
            max_timebtc = datetime.datetime.now().time()
        if float(conv(priceeth)) <= min(yeth):
            min_priceeth = priceeth
            min_timeeth = datetime.datetime.now().time()
        if float(conv(priceeth)) >= max(yeth):
            max_priceeth = priceeth
            max_timeeth = datetime.datetime.now().time()
    print('Change for bitcoin: ', float(ybtc[-1])-float(ybtc[0]))
    print('Percent change for bitcoin: ', (float(ybtc[-1])-float(ybtc[0]))/float(ybtc[0]))
    print('Change for ethereum: ', float(yeth[-1])-float(yeth[0]))
    print('Percent change for ethereum: ', (float(yeth[-1])-float(yeth[0]))/float(yeth[0]))
    print('Min price btc, time: ', min_pricebtc, min_timebtc)
    print('Max price btc, time: ' , max_pricebtc, max_timebtc)
    print('Min price eth, time: ', min_priceeth, min_timeeth)
    print('Max price eth, time: ' , max_priceeth, max_timeeth)
    
    plt.plot(x,ybtc, x, yeth)
    plt.show()
    again = input('Back to main menu?(y/n)')
    if again == 'y':
        main()
    

def main():
    x = int(input('Which mode?'))
    if x == 1:
        mode1()
    else:
        mode2()
main()


```
